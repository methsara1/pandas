```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Load the dataset
data = pd.read_csv('w2_position_salaries.csv')

# Separate the input (X) and output (y) variables
X = data['Level'].values.reshape(-1, 1)
y = data['Salary'].values

# Train a simple linear regression model
linear_reg = LinearRegression()
linear_reg.fit(X, y)
linear_r2 = r2_score(y, linear_reg.predict(X))

# Train polynomial regression models with degrees 2, 3, and 4
r2_scores = []
for degree in range(2, 5):
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(X)
    poly_reg = LinearRegression()
    poly_reg.fit(X_poly, y)
    r2_scores.append(r2_score(y, poly_reg.predict(X_poly)))

# Print the R2 scores
print("R2 score for simple linear regression:", linear_r2)
print("R2 scores for polynomial regression:")
for degree, r2_score in zip(range(2, 5), r2_scores):
    print(f"Degree {degree}: {r2_score}")
```
